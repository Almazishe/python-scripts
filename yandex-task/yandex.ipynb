```python
import pandas as pd
```

```python
df = pd.read_csv("data.csv")
df = df.filter(items=['buy_ts', 'uid', 'revenue'])
df_june = df.loc[df['buy_ts'].str.find(".06.2016") != -1]
unique_uids = df_june.uid.unique()

print("June", len(unique_uids))
```

```python
months = {
    1: "Jan",
    2: "Feb",
    3: "Mar",
    4: "Apr",
    5: "May",
    6: "Jun",
    7: "Jul",
    8: "Aug",
    9: "Sep",
    10: "Oct",
    11: "Nov",
    12: "Dec"
}
```

```python
m = 6
y = 2016

month = 6
year = 2016

for j in range(12):
    df_unique = df.loc[df['buy_ts'].str.find(f".{m:02}.{y}") != -1]
    unique_uids = df_unique.uid.unique()
    month = m
    year = y

    print(f"STARTING {months[m]} {y}")
    for i in range(13):
        df_month = df.loc[df['buy_ts'].str.find(f".{month:02}.{year}") != -1]
        df_month = df_month.loc[df_month['uid'].isin(unique_uids)]
        unique_uids = df_month.uid.unique()

        revenues = df_month.revenue.values
        
        sum_of_revenues = sum(revenues)

        print(f"{months[month]} {year} -", len(unique_uids), "-", sum_of_revenues)
        month += 1
        if month == 13:
            month = 1
            year += 1

    m += 1
    if m == 13:
        m = 1
        y += 1
    print("\n\n")
```
