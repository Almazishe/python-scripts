```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read data, and append all data to 1 dataframe
df1 = pd.read_csv("apartments_1.csv")
df2 = pd.read_csv("apartments_2.csv")
df3 = pd.read_csv("apartments_3.csv")
df4 = pd.read_csv("apartments_4.csv")
df5 = pd.read_csv("apartments_5.csv")
df6 = pd.read_csv("apartments_6.csv")

df = df1.append(df2)
df = df.append(df3)
df = df.append(df4)
df = df.append(df5)
df = df.append(df6)

# Check NaN values
df.isna().sum()
```

```python
# Sort all values by Link
df.sort_values("Link", inplace = True)
full_count = len(df)
full_count
```

```python
# dropping ALL duplicte apartments
df.drop_duplicates(subset ="Link",
                     keep = False, inplace = True)
without_duplicates = len(df)
without_duplicates
```

```python
# Number of duplicate apartments
full_count - without_duplicates
```

```python
# Get most frequent area and replace with it all NaN values
most_frequent_area = df.Area.mode()[0]
df.Area = df.Area.fillna(most_frequent_area)
df.isna().sum()
```

```python
# Get most frequent number of rooms and replace with it all NaN values
most_frequent_num_of_room = df.RoomNum.mode()[0]
df.RoomNum = df.RoomNum.fillna(most_frequent_num_of_room)
df.isna().sum()
```

```python
# Get most frequent city and replace with it all NaN values
most_frequent_city = df.City.mode()[0]
df.City = df.City.fillna(most_frequent_city)
df.isna().sum()
```

```python
# Get most frequent price and replace with it all NaN values
most_frequent_price = df.Price.mode()[0]
df.Price = df.Price.fillna(most_frequent_price)
df.isna().sum()
```

```python
# Get most frequent year of house and replace with it all NaN values
most_frequent_year = df.HouseYear.mode()[0]
df.HouseYear = df.HouseYear.fillna(most_frequent_year)
df.isna().sum()
```

```python
# Get unique city names
cities = np.unique(df['City'])
cities
```

```python
# Get unique House Years
years = np.unique(df['HouseYear'])
years
```

```python
# Get unique apartment areas
areas = np.unique(df['Area'])
areas
```

```python
# Get unique number of rooms
room_nums = np.unique(df['RoomNum'])
room_nums
```

```python
# Get unique prices
prices = np.unique(df['Price'])
prices
```

```python
for i, city in enumerate(cities):
    plt.scatter('HouseYear', 'Area',
                data=df.loc[df.City==city, :],
                s=1, label=str(city))
plt.title("Scatterplot of Apartments Area vs Year of building", fontsize=22)
```
